## Einbinden der interaktiven Schnittstelle

Im folgenden wird ein Notebook erstellt welches über nbinteract in eine interaktive HTML Seite umgewandelt werden kann.

```python
#nbi:hide_code

from ipywidgets import interact

import nbinteract as nbi

from numpy import zeros

import matplotlib.pyplot as plt 

%matplotlib inline
```

```python
def plotnew(funktion, number):
    x=zeros(number*100)
    y=zeros(number*100)
    for i in range(0,number*100):
        x[i]=i/100
        if funktion=='x^2':
            y[i]=x[i]**2
    plt.plot(x,y)    
    return '{} wird mit x von 0 bis {} geplottet!'.format(funktion, number)
```

```python
interact(plotnew, funktion='x^2', number=(0, 10));
```
